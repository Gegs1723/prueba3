```python
import random
import datetime
import string
import random

# Definir la longitud de la cadena aleatoria
longitud = 12

# Generar una cadena aleatoria de números y letras
caracteres = string.ascii_letters + string.digits
cod_veri = ''.join(random.choice(caracteres) for _ in range(longitud))

# Obtener la fecha actual
fecha_actual = datetime.date.today()

# Mostrar la fecha en formato personalizado
fecha_formateada = fecha_actual.strftime("%d/%m/%Y")

folio = random.randint(100000, 999999)
incrp = random.randint(1, 10000000)

from IPython.display import clear_output

def validar_rut(rut):
    rut = rut.replace(".", "").replace("-", "")

    if len(rut) != 9 or not rut[:-1].isdigit():
        return False

    multiplicadores = [2, 3, 4, 5, 6, 7, 2, 3]
    total = sum(int(digito) * multiplicador for digito, multiplicador in zip(rut[:-1][::-1], multiplicadores))
    resto = total % 11
    digito_verificador = str(11 - resto)

    if digito_verificador == "11":
        digito_verificador = "0"
    elif digito_verificador == "10":
        digito_verificador = "K"

    return rut[-1].upper() == digito_verificador

def grabar_persona():
    clear_output()
    print("Bienvenido al programa de TOTEM de autoservicio de atención rápida")
    print("\topcion.- crear (GRABAR) un registro de datos \n")
    print("comencemos con el ingreso de datos \n")
    rut = input("RUT (formato: XX.XXX.XXX-X) : ")
    while not validar_rut(rut):
        print("RUT inválido. Por favor, ingréselo nuevamente.")
        rut = input("RUT (formato: XX.XXX.XXX-X) : ")

    nombre = input("Nombre                      : ")
    while len(nombre) < 8:
        print("El nombre debe tener al menos 8 caracteres.")
        nombre = input("Ingrese el nombre de la persona: ")

    edad = int(input("Edad                        : "))
    while edad < 0:
        print("La edad debe ser mayor o igual a 0.")
        edad = int(input("Ingrese la edad de la persona: "))

    pais_nacimiento = input("País de nacimiento          : ")
    ciudad_nacimiento = input("Ciudad de nacimiento        : ")
    sexo= input("Sexo        :")

    # Mostrar los datos ingresados
    clear_output()
    print("\nDatos ingresados:\n")
    print("RUT                 :", rut)
    print("Nombre              :", nombre)
    print("Edad                :", edad)
    print("País de nacimiento  :", pais_nacimiento)
    print("Ciudad de nacimiento:", ciudad_nacimiento)
    print("sexo                :" , sexo)

    # Preguntar si los datos son correctos
    respuesta = input("\n¿Son correctos los datos? (s/n): ")
    if respuesta.lower() == "s":
        persona = {
            "rut": rut,
            "nombre": nombre,
            "edad": edad,
            "pais_nacimiento": pais_nacimiento,
            "ciudad_nacimiento": ciudad_nacimiento,
            "sexo" : sexo
        }
        print("Datos de la persona grabados correctamente.")
    else:
        print("Se canceló el proceso de grabación. Puede volver a ingresar los datos.")
        persona = None

    return persona

def buscar_persona(personas):
    clear_output()
    print("Bienvenido al programa de TOTEM de autoservicio de atención rápida")
    print("\tEsta en la opción de BUSQUEDA de personas\n")
    rut = input("Ingrese el RUT a buscar (formato: XX.XXX.XXX-X): ")
    for persona in personas:
        if persona["rut"] == rut:
            print("\nInformación de la persona:\n")
            print(f"RUT                 : {persona['rut']}")
            print(f"Nombre              : {persona['nombre']}")
            print(f"Edad                : {persona['edad']}")
            print(f"País de nacimiento  : {persona['pais_nacimiento']}")
            print(f"Ciudad de nacimiento: {persona['ciudad_nacimiento']}")
            print(f"sexo                : {persona['sexo']}")
            if persona['pais_nacimiento'].lower() == 'chile':
                print(f"\n{persona['nombre']} es de Chile.")
            else:
                print(f"\n{persona['nombre']} es extranjero(a)")
            return

    print("No se encontró ninguna persona con ese RUT.")

def imprimir_certificados(personas):
    clear_output()
    print("Bienvenido al programa de TOTEM de autoservicio de atención rápida")
    print("\tEsta en la opción para Imprimir certificados \n")

    print("1.- Certificado Nacimiento")
    print("2.- Certificado de matrimonio")
    print("3.- Certificado de residencia")

    opcion = input("\nIngrese el número de certificado que desea imprimir: ")

    if opcion.isdigit() and 1 <= int(opcion) <= 3:
        certificado = ""
        if opcion == "1":
            print("\nIngrese algunos datos para completar el certificado")
            fecha_nacimiento = input("\nFecha de nacimiento (DD/MM/AAAA): ")
            nombre_padre = input("Nombre del padre: ")
            nombre_madre = input("Nombre de la madre: ")
            rut = input("\nIngrese el RUT a buscar (formato: XX.XXX.XXX-X):")
            fecha_inicial = datetime.datetime.strptime(fecha_nacimiento, "%d/%m/%Y").date() # Obtener la fecha inicial ingresada por el usuario
            año_inicial = fecha_inicial.year # Obtener el año de la fecha inicial
            año_actual = datetime.date.today().year # Obtener el año actual
            año_aleatorio = random.randint(año_inicial, año_actual) # Generar un año aleatorio entre el año inicial y el año actual
            for persona in personas:
                if persona['rut'] == rut:
                    clear_output()
                    hora = random.randint(0, 23)
                    minuto = random.randint(0, 59)
                    hora_nac= "{:02d}:{:02d}".format(hora, minuto)
                    print("SERVICIO DE REGISTRO\nCIVIL E IDENTIFICACION")
                    print("\t\t\t\t\t\t\tFOLIO: ", folio)
                    print("\t\t\t\t\t\t\tCodigo  Verificacion :")
                    print("\t\t\t\t\t\t\t\t",cod_veri)
                    print("\n------------------------------ Certificado de Nacimiento ------------------------------")
                    print("\n")
                    print(f"RUT                      : {persona['rut']}")
                    print(f"Nombre                   : {persona['nombre']}")
                    print(f"Edad                     : {persona['edad']}")
                    print(f"País de nacimiento       : {persona['pais_nacimiento']}")
                    print(f"Ciudad de nacimiento     : {persona['ciudad_nacimiento']}")
                    print(f"Fecha de nacimiento      :", fecha_nacimiento)
                    print(f"hora de nacimiento       :",hora_nac )
                    print(f"Nombre del padre         :", nombre_padre)
                    print(f"Nombre de la madre       :", nombre_madre)
                    print(f"Sexo                     : {persona['sexo']}")
                    print(f"Fecha de inscripción     :", fecha_formateada)
                    print(f"Año del libro de registro:", año_aleatorio)
                    print("-----------------------------------------------------------------------------------------")
                    print()
                    return

        elif opcion == "2":
            print("\nIngrese algunos datos para completar el certificado")
            fecha_nacimiento = input("\nFecha de nacimiento (DD/MM/AAAA): ")
            circuncrp =      input("cicunscripcion         :")
            nombre_conjuge = input("Nombre del(a) conyugue :")
            fec_na_conjuge = input("fecha nac. conyugue    :")
            rut_conyugue   = input("RUT conyugue           :")
            rut = input("\nIngrese el RUT a buscar (formato: XX.XXX.XXX-X):")
            fecha_inicial = datetime.datetime.strptime(fecha_nacimiento, "%d/%m/%Y").date() # Obtener la fecha inicial ingresada por el usuario
            año_inicial = fecha_inicial.year # Obtener el año de la fecha inicial
            año_actual = datetime.date.today().year # Obtener el año actual
            año_aleatorio = random.randint(año_inicial, año_actual) # Generar un año aleatorio entre el año inicial y el año actual
            for persona in personas:
                if persona['rut'] == rut:
                    clear_output()
                    hora = random.randint(0, 23)
                    minuto = random.randint(0, 59)
                    hora_nac= "{:02d}:{:02d}".format(hora, minuto)
                    print("SERVICIO DE REGISTRO\nCIVIL E IDENTIFICACION")
                    print("\t\t\t\t\t\t\tFOLIO: ", folio)
                    print("\t\t\t\t\t\t\tCodigo  Verificacion :")
                    print("\t\t\t\t\t\t\t\t",cod_veri)
                    print("\n------------------------------ Certificado de Matrimonio ------------------------------")
                    print("cicunscripcion         :", circuncrp)
                    print("numero de incripcion   :", incrp)
                    print("")
                    print("Nombre del(a) conyugue :",nombre_conjuge )
                    print("RUT conyugue           :",rut_conyugue)
                    print("fecha nac. conyugue    :",fec_na_conjuge)
                    print("")
                    print("")
                    print(f"RUT                 : {persona['rut']}")
                    print(f"Nombre              : {persona['nombre']}")
                    print(f"fecha de nacimiento :", fecha_nacimiento)
                    print("-----------------------------------------------------------------------------------------")

        elif opcion == "3":
            rut = input("\nIngrese el RUT a buscar (formato: XX.XXX.XXX-X):")
            dire = input("ingrese su direccion                            :")
            for persona in personas:
                if persona['rut'] == rut:
                                clear_output()
                                hora = random.randint(0, 23)
                                minuto = random.randint(0, 59)
                                hora_nac= "{:02d}:{:02d}".format(hora, minuto)
                                print("SERVICIO DE REGISTRO\nCIVIL E IDENTIFICACION")
                                print("\t\t\t\t\t\t\tFOLIO: ", folio)
                                print("\t\t\t\t\t\t\tCodigo  Verificacion :")
                                print("\t\t\t\t\t\t\t\t",cod_veri)
                                print("\n----- Certificado de Residencia -----")
                                print("\nDatos de la persona encontrada:\n")
                                print(f"RUT                 : {persona['rut']}")
                                print(f"Nombre              : {persona['nombre']}")
                                print(f"Edad                : {persona['edad']}")
                                print(f"País de nacimiento  : {persona['pais_nacimiento']}")
                                print(f"Ciudad de nacimiento: {persona['ciudad_nacimiento']}")
                                print(f"Direccion          :", dire)

                                if persona['pais_nacimiento'].lower() == 'chile':
                                    print(f"\n{persona['nombre']} es de Chile.")
                                else:
                                    print(f"\n{persona['nombre']} es extranjero(a)")
                                print("----------------------------------------------")
                                print()
                                return

            print("No se encontró ninguna persona con ese RUT.")
        else:
            print("Opción inválida.")

def eliminar_persona(personas):
    clear_output()
    print("Bienvenido al programa de TOTEM de autoservicio de atención rápida")
    print("Esta en la opción de eliminar registro de una persona\n")
    rut = input("Ingrese el RUT de la persona a eliminar: ")
    for persona in personas:
        if persona["rut"] == rut:
            personas.remove(persona)
            print("La persona ha sido eliminada correctamente.")
            return
    print("No se encontró ninguna persona con ese RUT.")

def imprimir_menu():
    print("\n------ MENÚ ------")
    print("1. Grabar")
    print("2. Buscar")
    print("3. Imprimir certificados")
    print("4. Eliminar")
    print("5. Salir")
    print("6. Mostrar personas guardadas")

def main():
    print("Bienvenido al programa de TOTEM de autoservicio de atención rápida")
    print("Versión 1.2\n")
    personas = []

    while True:
        imprimir_menu()
        opcion = input("Ingrese el número de la opción que desea seleccionar: ")

        if opcion == "1":
            persona = grabar_persona()
            if persona:
                personas.append(persona)
        elif opcion == "2":
            buscar_persona(personas)
        elif opcion == "3":
            imprimir_certificados(personas)
        elif opcion == "4":
            eliminar_persona(personas)
        elif opcion == "5":
            clear_output()
            print("\nGracias por utilizar el  programa de TOTEM de autoservicio de atención rápida.\nEsta es la version 1.0 del programa \nfue realiado por Gabriel \nHasta luego.")
            break
        elif opcion == "6":
            print("Personas guardadas:")
            for persona in personas:
                print(f"RUT: {persona['rut']}, Nombre: {persona['nombre']}")
        else:
            print("Opción inválida. Por favor, ingrese una opción válida.")

if __name__ == "__main__":
    main()
```
